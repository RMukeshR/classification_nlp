```python
import numpy as np
def convert_embedding(embedding_str):
    embedding_list = embedding_str.strip('[]').split()
    return np.array(embedding_list, dtype=np.float32)

df['w2v_embedding'] = df['w2v_embedding'].apply(convert_embedding)
```

```python
for i in range(0,1):
    print(df_sel["w2v_embedding"][i])
    print(type(df_sel["w2v_embedding"][i]))
    arr = np.array(df_sel["w2v_embedding"][i])
    print(arr)
    print(type(arr))
```

```python
import numpy as np

# Convert w2v_embedding strings to actual arrays
df_sel['w2v_embedding'] = df_sel['w2v_embedding'].apply(lambda x: np.fromstring(x.strip("[]"), sep=' '))
```

```python
df_sel['w2v_embedding'][0]
```

```python
X = np.array(df_sel['w2v_embedding'].tolist())  

X.shape
```

```python
y = df['level']
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=69)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score

results = {}


model = SVC(kernel="linear")
model.fit(X_train, y_train)
y_pred = model.predict(X_test)


accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred, output_dict=True)

results["linear"] = {
    'accuracy': accuracy,
    'precision': report['weighted avg']['precision'],
    'recall': report['weighted avg']['recall'],
    'f1-score': report['weighted avg']['f1-score']
}
```

```python
results_df = pd.DataFrame(results).T
print(results_df)
```
