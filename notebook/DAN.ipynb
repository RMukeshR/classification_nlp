```python
import numpy as np
def convert_embedding(embedding_str):
    embedding_list = embedding_str.strip('[]').split()
    return np.array(embedding_list, dtype=np.float32)

```

```python
df['w2v_embedding'] = df['w2v_embedding'].apply(convert_embedding)
```
